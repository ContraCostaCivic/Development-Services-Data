### Desk Workload
Here we try to look at whether the workload for desk workers is dependent on day of the week and if resources should be allocated differently on different days

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./front_counter_cleaned.csv")
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
#Get info on day of week and month (we don't use month but may be helpful for further analysis)
df["weekday"] = df["date"].dt.weekday_name
df["month"] = df["date"].dt.month
```

```python
#Calculate at average number of visitors by day
smalldf = df[["date", "weekday", "n_visitors"]]
smalldf = smalldf[df["date"] >= "2018-01-01"] #We have a lot of data but older data might not be as relevant
smalldf = smalldf.dropna()
#smalldf = smalldf[smalldf["n_visitors"].str.isnumeric()]
smalldf["n_visitors"] = smalldf.n_visitors.astype(int)

#can probably do this in one line
smalldf = smalldf[smalldf["weekday"] != "Saturday"]
smalldf = smalldf[smalldf["weekday"] != "Sunday"]
smalldf = smalldf[smalldf["weekday"] != "Friday"]
visitors_table = smalldf.groupby(["weekday"]).mean()
```

```python
#Calculate at average inspection requests by day
inspecdf = df[["date", "weekday", "inspection_requests"]]
inspecdf = inspecdf.dropna()
inspecdf = inspecdf[inspecdf["inspection_requests"].str.isnumeric()]
inspecdf["inspection_requests"] = inspecdf.inspection_requests.astype(int)

#can probably do this in one line
inspecdf = inspecdf[inspecdf["weekday"] != "Saturday"]
inspecdf = inspecdf[inspecdf["weekday"] != "Sunday"]
inspecdf = inspecdf[inspecdf["weekday"] != "Friday"]
inspect_table = inspecdf.groupby(["weekday"]).mean()
```

```python
inspect_table
```

```python
#fix indicies and then move back to columns
inspect_table = inspect_table.reindex(["Monday", "Tuesday", "Wednesday", "Thursday"])
visitors_table = visitors_table.reindex(["Monday", "Tuesday", "Wednesday", "Thursday"])
inspect_table = inspect_table.reset_index()
visitors_table = visitors_table.reset_index()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

#Plot Average Daily Count of Inspection Requests by Day of Week
ax = sns.barplot(x = "weekday", y = "inspection_requests", data=inspect_table)
ax.set_title('Average Number of Inspection Requests by Day of the Week')
ax.set_ylabel('Inspection Requests')
ax.set_xlabel('Day of the Week')
ax.figure.savefig("Inspection Request Count.png") #save as image
```

```python
#Plot Average Daily Visitor Count by Day of Week
ax = sns.barplot(x = "weekday", y = "n_visitors", data=visitors_table)
ax.set_title('Average Number of Visitors by Day of the Week')
ax.set_ylabel('Number of Visitors')
ax.set_xlabel('Day of the Week')
ax.figure.savefig("Visitor Count.png") #save as image
```
